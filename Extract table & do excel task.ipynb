### Pivot table

```python
import pandas as pd
```

```python
df=pd.read_excel(r'C:\Users\Admin\Desktop\Automation\supermarket_sales.xlsx')
```

```python
print(df)
```

```python
df1=df[['Gender','Product line','Total']]
print(df1)
```

```python
pivot_table=df1.pivot_table(index='Gender',columns='Product line',values='Total',aggfunc='sum')
print(pivot_table)
```

```python
pivot_table.to_excel('pivot_table.xlsx','Report',startrow=4)
```

### add barchart

```python
from openpyxl import load_workbook
```

```python
wb=load_workbook('pivot_table.xlsx')
sheet=wb['Report']

min_col=wb.active.min_column
max_col=wb.active.max_column
min_row=wb.active.min_row
max_row=wb.active.max_row


print(min_col)
print(max_col)
print(min_row)
print(max_row)
```

```python
from openpyxl.chart import BarChart, Reference
```

```python
barchart=BarChart()

data=Reference(sheet,min_col=min_col+1,max_col=max_col,min_row=min_row,max_row=max_row)
categories=Reference(sheet,min_col=min_col+1,max_col=min_col,min_row=min_row+1,max_row=max_row)

barchart.add_data(data,titles_from_data=True)
barchart.set_categories(categories)

sheet.add_chart(barchart,"B12")

barchart.title="Sales by product"
barchart.style=5
wb.save('barchart.xlsx')
```

### Excel formula in Python

```python
from openpyxl import load_workbook
from openpyxl.utils import get_column_letter
```

```python
WB=load_workbook('barchart.xlsx')
sheet=WB['Report']
```

```python
mincolumn=WB.active.min_column
maxcolumn=WB.active.max_column
minrow=WB.active.min_row
maxrow=WB.active.max_row

for i in range (mincolumn+1,maxcolumn+1):
    letter=get_column_letter(i)
    sheet[f'{letter}{maxrow+1}']=f'=SUM({letter}{minrow+1}:{letter}{maxrow})'
          
WB.save('report.xlsx')
```
